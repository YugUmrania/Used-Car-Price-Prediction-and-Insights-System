```python
luxury_brand = [
    "Mercedes-Benz",
    "BMW",
    "Audi",
    "Land Rover",
    "Volvo",
    "Jaguar",
    "Mini",
    "Porsche",
    "Lexus",
    "Maserati",
    "Bentley",
    "Lamborghini",
    "Rolls-Royce",
    "Jeep",
    "Hummer"
]

def extract_Brand_model(car_name, brand_list = luxury_brand):
    if not isinstance(car_name, str):
        return None, None
    
    car_name = car_name.strip()
    
    # Sort brands by length
    brand_list = sorted(brand_list, key=len, reverse=True)
    
    for brand in brand_list:
        if car_name.lower().startswith(brand.lower()):
            model = car_name[len(brand):].strip()
            return brand, model
        
        
    parts = car_name.split(" ", 1)
    brand = parts[0]
    model = parts[1] if len(parts) > 1 else ""
    
    return brand, model
    
```

```python
# Universal Process function

def process_model(row, luxury_brands):
    brand = row['Brand']
    model = row['Model']
    
    if pd.isna(brand) and pd.isna(row):
        return ""
    
    brand_norm = brand.lower().strip()
    model_norm = model.lower().strip()
    
    if brand_norm in luxury_brands:
        
        brand_models = luxury_model_map.get(brand_norm, {})

        for base_model, patterns in brand_models.items():
            for p in patterns:
                if p in model_norm:
                    return base_model.title()

        # If luxury brand but model not mapped
        return model.title()

    # Non-luxury → keep original model
    return model
```

```python
# Applying that function to the model column

dataset_cardekho_final.loc[:, "Final_Model"] = (
    dataset_cardekho_final.apply(lambda row: process_model(row, luxury_brands), axis=1)
)
```

```python
# Part 1: Define EXACT luxury brands + REQUIRED COUNTS

luxury_brand_limits = {
    "audi": 60,
    "bmw": 39,
    "mercedes-benz": 35,
    "jaguar": 6,
    "land rover": 5,
    "volvo": 4,
    "jeep": 3
}

# Part 2: Filter ONLY these luxury brands from CarDekho

luxury_cardekho = dataset_cardekho_final[
    dataset_cardekho_final["Brand"].str.lower().isin(luxury_brand_limits.keys())
].copy()

luxury_cardekho["Brand"] = luxury_cardekho["Brand"].str.lower()

# PART 3: Select ONLY the required number of cars per brand

selected_luxury_cars = []

for brand, limit in luxury_brand_limits.items():
    brand_data = luxury_cardekho[luxury_cardekho["Brand"] == brand]
    
    # Safety check
    if len(brand_data) < limit:
        print(f"⚠ Warning: {brand} has only {len(brand_data)} cars")
        selected = brand_data
    else:
        selected = brand_data.sample(n=limit, random_state=42)
    
    selected_luxury_cars.append(selected)

luxury_final = pd.concat(selected_luxury_cars, ignore_index=True)
```

```python
# For verification let's check the value counts in the luxury_final

luxury_final["Brand"].value_counts()
```

```python
# Numeric Columns

numeric_cols = ["Price", "kmDriven", "Age"]

# Part 4: METHOD USED: IQR (Best for skewed automobile data)

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    return df[(df[column] >= lower) & (df[column] <= upper)]

# Apply outlier removal per brand (VERY IMPORTANT)

clean_luxury = []

for brand in luxury_final["Brand"].unique():
    brand_df = luxury_final[luxury_final["Brand"] == brand]
    
    for col in numeric_cols:
        brand_df = remove_outliers_iqr(brand_df, col)
    
    clean_luxury.append(brand_df)

luxury_cleaned = pd.concat(clean_luxury, ignore_index=True)
```

```python
# Fix year outliers and add age + km_per_year
df_model = New_Final_Cleaned_df_copy.copy()

df_model = df_model[(df_model['year'] >= 1995) & (df_model['year'] <= 2026)].copy()
df_model['age'] = 2026 - df_model['year']
df_model['km_per_year'] = df_model['kmdriven'] / (df_model['age'] + 1)  # Adding 1 to avoid division by zero
df_model.drop(columns=['year'], inplace=True)

# Also changing the position of the age and km_per_year column to be after the kmdriven column for better readability.
col = df_model.pop('age')
col1 = df_model.pop('km_per_year')
df_model.insert(df_model.columns.get_loc('model') + 1, 'age', col)
df_model.insert(df_model.columns.get_loc('kmdriven') + 1, 'km_per_year', col1)

df_model.reset_index(drop=True, inplace=True)
print(df_model.shape)
df_model.head()
```

```python
# Starting with Baseline Models

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error, mean_absolute_percentage_error
import numpy as np

def evaluate(name, model):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    
    # # Convert back from log scale to actual Indian Rupees
    # actual      = np.expm1(y_test)
    # predicted   = np.expm1(preds)
    
    # # All Metrics
    # r2 = r2_score(y_test, preds)    #  on log scale (standard)
    # mae = mean_absolute_error(actual, predicted)    # ₹ Rupees
    # mse = mean_squared_error(actual, predicted)     # # ₹^2
    # rmse = np.sqrt(mse)
    # mape = mean_absolute_percentage_error(actual, predicted) * 100  # %
    
    
    # All Metrics
    r2   = r2_score(y_test, preds)
    mae  = mean_absolute_error(y_test, preds)
    mse  = mean_squared_error(y_test, preds)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_test, preds) * 100
    
    
    print(f"\n  Model  : {name}")
    print(f"  {'─'*35}")
    print(f"  R²     : {r2:.4f}")
    print(f"  MAE    : {mae:.4f}")
    print(f"  MSE    : {mse:.4f}")
    print(f"  RMSE   : {rmse:.4f}")
    print(f"  MAPE   : {mape:.2f}%")
    print(f"  {'─'*35}")
    
    return model

lr = evaluate("Linear Regression", LinearRegression())
rf = evaluate("Random Forest Regressor", RandomForestRegressor(n_estimators=100, random_state=42))
xgboost = evaluate("XGBoost ", XGBRegressor(n_estimators=300, learning_rate=0.05,
                                        max_depth=6, subsample=0.8, 
                                        colsample_bytree=0.8, random_state=42))
```

```python
# Target Encoding piepline
from category_encoders import TargetEncoder
from sklearn.model_selection import train_test_split

# Start fresh from cleaned df
df_te = New_Final_Cleaned_df.copy()

# Lowercase
cat_cols = df_te.select_dtypes(include='object').columns
df_te[cat_cols] = df_te[cat_cols].apply(lambda x: x.str.lower().str.strip())
df_te.columns = df_te.columns.str.lower()

# Fix + feature engineering
df_te['kmdriven'] = df_te['kmdriven'].replace(0, df_te['kmdriven'].median())
df_te['kmdriven'] = np.log1p(df_te['kmdriven'])
df_te['price']    = np.log1p(df_te['price'])
df_te = df_te[(df_te['year'] >= 1995) & (df_te['year'] <= 2026)].copy()
df_te['age']         = 2026 - df_te['year']
df_te['km_per_year'] = df_te['kmdriven'] / (df_te['age'] + 1)

df_te['log_km'] = np.log1p(df_te['kmdriven'])
df_te['age_squared'] = df_te['age'] ** 2

df_te.drop(columns=['year'], inplace=True)
    
# OHE for transmission, fueltype, owner
df_te = pd.get_dummies(df_te, columns=['transmission', 'fueltype', 'owner'], drop_first=False)
bool_cols = df_te.select_dtypes(include='bool').columns
df_te[bool_cols] = df_te[bool_cols].astype(int)

df_te.reset_index(drop=True, inplace=True)

# Remove extreme outliers
for col in ['price', 'age', 'km_per_year']:
    Q1 = df_te[col].quantile(0.25)
    Q3 = df_te[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_te = df_te[(df_te[col] >= lower) & (df_te[col] <= upper)]


# Split first (IMPORTANT — fit encoder only on train)
X_te = df_te.drop(columns=['price'])
y_te = df_te['price']

X_train_te, X_test_te, y_train_te, y_test_te = train_test_split(
    X_te, y_te, test_size=0.2, random_state=42
)

# Target encode brand and model
encoder = TargetEncoder(cols=['brand', 'model'])
X_train_te = encoder.fit_transform(X_train_te, y_train_te)
X_test_te  = encoder.transform(X_test_te)

print(X_train_te.shape)
```

```python
# Retain XGBoost with the Target Encoded data

best_params = {
    'subsample'        : 0.7,
    'n_estimators'     : 700,
    'min_child_weight' : 1,
    'max_depth'        : 6,
    'learning_rate'    : 0.05,
    'gamma'            : 0,
    'colsample_bytree' : 0.7
}

xgb_te = XGBRegressor(
    **best_params,
    random_state=42,
    early_stopping_rounds=50
)

xgb_te.fit(
    X_train_te,
    y_train_te,
    eval_set=[(X_test_te, y_test_te)],   # validation set
    verbose=False
)

preds_te = xgb_te.predict(X_test_te)

r2   = r2_score(y_test_te, preds_te)
mae  = mean_absolute_error(y_test_te, preds_te)
mse  = mean_squared_error(y_test_te, preds_te)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_test_te - preds_te) / y_test_te)) * 100

print(f"\n  Model  : XGBoost + Target Encoding")
print(f"  {'─'*35}")
print(f"  R²     : {r2:.4f}")
print(f"  MAE    : {mae:.4f}")
print(f"  MSE    : {mse:.4f}")
print(f"  RMSE   : {rmse:.4f}")
print(f"  MAPE   : {mape:.2f}%")
print(f"  {'─'*35}")
```
